## While example that is time based

Let's talk about food

```python
# Example of doing something while waiting. 
# Imagine you are working a job that lasts 30 seconds
from time import time
from random import choice
from random import randint

# Define our foods
fruits = ['apple', 'banana', 'orange', 'plum', 'apricot']
vegetables = ['carrot', 'brocoli']


def get_food():
    """Get a random food"""
    number = randint(0, 1)
    
    if number == 1:
        return choice(vegetables)
    else:
        return choice(fruits)

    
def is_vegetable(food):
    """Returns True if `food` is a vegetable"""
    return food in vegetables


def is_fruit(food):
    """Returns true if `food` is a fruit"""
    for fruit in fruits: 
        if fruit == food:
            return True
    
    return False


# Get the current time so we know when to stop patrolling
starting_time = time()
shift_length_in_seconds = 1
number_of_vegetables = 0

end_of_shift = starting_time + shift_length_in_seconds

# time() returns the current time in seconds
# see the doc for more details https://docs.python.org/3.5/library/time.html#time.time
while time() < end_of_shift:
    print('I\'m in the while loop')
    food = get_food()
    
    if is_vegetable(food):
        number_of_vegetables += 1
        print('Saved the world from a', food)

print('\n\nWe found', number_of_vegetables, 'vegetables')
    
```

```python
import image
import random
import sys

def get_base_offset():
    """"Returns the base offset to make the image blur"""
    # Increase this value to make the image more blury
    return 1

def create_image():
    """Gives the base luther college bell image"""
    return image.Image("luther.jpg")


def create_window(width, height):
    """Creates the window to draw on"""
    return image.ImageWin(width, height)


def get_random_offset():
    """Returns a random offset for finding pixel locations"""
    negative_base = get_base_offset() * -1
    return random.randint(negative_base, get_base_offset())


# Increase program run time so it can finish running
sys.setExecutionLimit(90000)

# Create the Image
img = create_image()
wn = create_window(img.getWidth(), img.getHeight())

# Fill in the window
img.draw(wn)


# Set the maximum offsets for setting the blur pixels
base_offset = get_base_offset()
max_rows = img.getHeight() - base_offset
max_columns = img.getWidth() - base_offset


for row in range(base_offset, max_rows):
    for col in range(base_offset, max_columns):
        
        col_offset = col + get_random_offset()
        row_offset = row + get_random_offset()
        
        existing_pixel = img.getPixel(col_offset, row_offset)
        
        red = existing_pixel.getRed()
        blue = existing_pixel.getGreen()
        green = existing_pixel.getBlue()
       
       
        newpix = image.Pixel(red, blue, green)
       
        img.setPixel(col, row, newpix)
```
